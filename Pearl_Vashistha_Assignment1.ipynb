Q6.Compare different delivery rating counts of different restaurants.
Also tell that which restaurants has got the highest rating counts.
Do this by creating a barplot

```python
name= list(data.Restaurant_Name)
rating= list(data.Delivery_Rating)
'''
'''

count =0
for i in name:
  count = count + 1
  index= count -1
  if type(i) != str:
    name[index]="NaN"

print(name)

# convert nan values to int so that plot

ser1= pd.Series(rating)
ser1 = ser1.fillna(0)
rating = list(ser1)

plt.bar(name, rating)
plt.xlabel("Restaurant Name")
plt.xticks(rotation=30)
plt.ylabel("Delivery Rating")
plt.title("Delhi NCR Restaurants", size= 10, color="blue")
plt.show()


```
